Part B: Data cleaning & preparation.

6. Handle missing value appropriately and explain your method in comments

```python
# Convert Salary to numeric and fill missing with median
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')
data['Salary'] = data['Salary'].fillna(data['Salary'].median())  # Median is robust

# Fill missing values in object columns with mode
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].fillna(data[col].mode()[0])  # Mode is most frequent value

# Check missing values
print(data.isnull().sum())
```

7. Check for duplicate records and remove them if present.

```python
# Before removing
print(data.shape)

# Remove duplicates
data.drop_duplicates(inplace=True)

# After removing
print(data.shape)
```

8. Convert relevent coloumns into correct data types(eg.,salary,experience).

```python
# Convert columns to numeric
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')
data['Salaries Reported'] = pd.to_numeric(data['Salaries Reported'], errors='coerce')

# Check data types
print(data.dtypes)
```

9. Create a new colomns categorizing experience levels(Beginner,Intermediate,Expert).

```python
# Create random Experience column
data['Experience'] = np.random.randint(0, 10, size=len(data))

# Function
def exp_level(exp):
    if exp < 2:
        return "Beginner"
    elif exp < 5:
        return "Intermediate"
    else:
        return "Expert"

# Apply function
data['Experience_Level'] = data['Experience'].apply(exp_level)

# Display output
print(data[['Experience', 'Experience_Level']].head())
```

10. Standardize or clean text columns if required(eg.,job roles,locations).

```python
# Clean column names first
data.columns = data.columns.str.strip().str.lower()

# Clean all text (object) columns
for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].str.strip().str.title()

# Check result
print(data.head())
```

18. Identify any outliers in salary distribution.

```python
# Ensure Salary is numeric
data['salary'] = pd.to_numeric(data['salary'], errors='coerce')

# Calculate Q1 and Q3
Q1 = data['salary'].quantile(0.25)
Q3 = data['salary'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier condition
outliers = data[(data['salary'] < Q1 - 1.5 * IQR) |
                (data['salary'] > Q3 + 1.5 * IQR)]

# Print outliers
print("Outliers:")
print(outliers)
```

23. Compare NumPy result with Pandas results.

```python
# Convert salary column to NumPy array (drop NaN)
salary_array = data['salary'].dropna().to_numpy()

# NumPy calculations
np_mean = np.mean(salary_array)
np_std = np.std(salary_array)
np_min = np.min(salary_array)
np_max = np.max(salary_array)

# Pandas calculations
pd_mean = data['salary'].mean()
pd_std = data['salary'].std()
pd_min = data['salary'].min()
pd_max = data['salary'].max()

# Print comparison
print("NumPy Mean:", np_mean)
print("Pandas Mean:", pd_mean)
print("\nNumPy Std Dev:", np_std)
print("Pandas Std Dev:", pd_std)
print("\nNumPy Min:", np_min)
print("Pandas Min:", pd_min)
print("\nNumPy Max:", np_max)
print("Pandas Max:", pd_max)
```
